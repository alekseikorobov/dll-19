```python
def get_words(path,size=(4,5),count=100):
    res = set()
    with open(path,'r') as f:
        for line in f.readlines():
            text = line.strip().strip('\t').replace('\n','').lower()            
            if len(text)< size[0]:
                continue
            for word in text.split():
                if len(word)>=size[0] and len(word)<=size[1]:
                    res.add(word)
                    
                if len(res)>=count:
                    return list(res)            
    return list(res)
            
words = get_words('data/onegin.txt')
print(words)
```

```python
from util.augmentation import Augmentation
from dataset.MyDataset import MyDataset
import matplotlib.pyplot as plt
import cv2
import numpy as np
from util.my_utils import get_center_loss,get_center_counters_fact,get_center_counters_predict
%matplotlib inline
#plt.plot([1])

input_size = 640
means = (0.485, 0.456, 0.406)
stds = (0.229, 0.224, 0.225)
num_workers = 1
net = 'resnet50'
is_training = False

conf = [{
    'fonts':['dataset/MyGenerator/font.ttf']
    , 'texts':[f'''
               select *
               from [table]
               ''']
    , 'text_colors':['#000000'] #,'#ff0000'
    , 'size_images':[(300,300)]
    , 'position_texts':[(50,12)] #,
    , 'font_sizes':[14]
    , 'background_colors':['#ffffff']
    , 'use_box':[False]
    , 'box_colors':['#555555']
    #, 'use_lines':['','|','-','\\','/']
    , 'use_lines':['0000']
    , 'lines_colors':['#555555']
    , 'box_colors':['#555555']
    # , 'use_special_chars':[True, False]
}]
dateset = MyDataset(conf,is_training=True,transform=Augmentation(size=input_size, mean=means, std=stds))

for d in dateset:
    image = d[0]
    img = image.permute(1,2,0).cpu().numpy()
    
    im_show0 = img.copy()
       
    centers_fact = get_center_counters_fact(d)    
    
    for cx,cy in centers_fact:
        cx,cy = int(cx.item()),int(cy.item())
        cv2.circle(im_show0, (cx, cy), 2, (255, 0, 00), 2)
    
    contours = d[6].cpu().numpy()
    cv2.drawContours(im_show0, contours.astype(np.int32), -1, (0, 0, 255), 2)
    plt.figure(figsize=(7,7))    
    plt.imshow(im_show0)
    plt.show()
    break

words = dateset.torch_text_dict.get_labels(d[9])

im_show0 = img.copy()
for contour, word in zip(contours, words):
    x, y, w, h = cv2.boundingRect(contour)
    im_part = im_show0[y:y+h, x:x+w]
    plt.title(word)
    plt.imshow(im_part)
    plt.show()
```
