## Задача распознования текста

Модель - **crnn**

- https://github.com/bgshih/crnn реализация 1
- https://github.com/meijieru/crnn.pytorch  реализация 2
- https://arxiv.org/abs/1507.05717 https://arxiv.org/pdf/1507.05717.pdf статья от 2015 года

- https://www.cs.toronto.edu/~graves/icml_2006.pdf

Реализация тут
- https://deepayan137.github.io/blog/markdown/2020/08/29/building-ocr.html#the-ctc-loss
